```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Generate a range of values
x = np.linspace(-4, 4, 1000)

# Calculate the PDF of the standard normal distribution
pdf = stats.norm.pdf(x)
```

```python
# Plot the PDF
plt.plot(x, pdf, label='Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Theoretical Null Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
# Create the plot
trace = go.Scatter(
    x=x,
    y=pdf,
    mode='lines',
    name='Standard Normal Distribution'
)

layout = go.Layout(
    title='Theoretical Null Distribution',
    xaxis=dict(title='Value'),
    yaxis=dict(title='Probability Density'),
    showlegend=True
)

fig = go.Figure(data=[trace], layout=layout)

# Display the plot
pyo.iplot(fig)
```

```python
# Parameters for the t-distribution
df = 2 # degrees of freedom

# Generate a range of values
x = np.linspace(-4, 4, 1000)

# Calculate the PDF of the t-distribution
pdf = stats.t.pdf(x, df)

# Create the plot
trace = go.Scatter(
    x=x,
    y=pdf,
    mode='lines',
    name=f't-distribution (df={df})'
)

layout = go.Layout(
    title='T-Distribution',
    xaxis=dict(title='Value'),
    yaxis=dict(title='Probability Density'),
    showlegend=True
)

fig = go.Figure(data=[trace], layout=layout)

# Display the plot
pyo.iplot(fig)
```
